# Loss Function Practise 

I am going to generate a dataset for two-feature linear regression using the datasets module from scikit-learn. The default data generated by this module is centered, meaning that the intercept, b0 is already set to zero.

Exercise 1

The n_samples input parameter lets me set how many data points I would like —- I’ve set it to 100. n_features lets me specify the number of features I want, which would be 2 in this case.

The coef = True attribute gives me the coefficients generated by fitting a linear regression model. 

* Print the coefficients given by this model. (Remember that intercept b0 is zero as the data is centered - so only printing the best fit values of b1 and b2 here!)


```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets
data, y, coefficients = datasets.make_regression(n_samples = 100, n_features = 2, coef = True, random_state = 23)
x1 = data[:,0]
x2 = data[:,1]

# 1. Print the coefficients
print(coefficients)
```

Output: [77.30856619 18.87896326]

Exercise 2

Write a pythonic function to output the value of the loss function at each data point for all possible values of b1 and b2. Fill in the code for the loss_function() so that it outputs loss according to the formula for the loss function 



* Fill in the code to calculate the error variable inside the function.

```python
# 2. Loss Function
def loss_function(b1,b2,y,x1,x2):
    error = y - b1*x1 - b2*x2
    loss = np.mean(error**2)
    return loss
```
